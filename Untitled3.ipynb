```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
fonts=FontProperties(fname='Fonts\STSONG.TTF')

```

```python
datadf = pd.read_excel('data\调查问卷2.xls')
datadf.head()
```

```python
dataflatten=np.array(datadf.iloc[:,1::]).flatten()
dataflatten=pd.DataFrame({'value':dataflatten})
datafre=dataflatten.groupby(by=['value'])['value'].count()
datafre=pd.DataFrame({'Item':datafre.index,'Freq':datafre.values}).sort_values('Freq',ascending=0)
datafre.plot(kind='bar',figsize=(12,6),legend=None)
plt.title('选择出现的频次',fontproperties=fonts)
plt.xlabel('')
plt.ylabel('频次',fontproperties=fonts,size=12)
plt.xticks(range(len(datafre)),datafre['Item'],rotation=90,fontproperties=fonts,size=9)
plt.show()
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
datanew=np.array(datadf.iloc[:,1::])
oht=TransactionEncoder()
oht_ary=oht.fit(datanew).transform(datanew).astype('int')
df=pd.DataFrame(oht_ary,columns=oht.columns_)
df.head()
```

```python
df_fre=apriori(df,min_support=0.3,use_colnames=True)
df_fre['lenth']=df_fre['itemsets'].apply(lambda x:len(x))
print(df_fre.shape)
df_fre_len2=df_fre[df_fre['lenth']>1]
print(df_fre_len2.shape)
df_fre_len2.sample(5)
```

```python
rule1=association_rules(df_fre,metric='lift',min_threshold=1.1)
rule1['antelen']=rule1['antecedents'].apply(lambda x:len(x))
rule1
```

```python
rule1[(rule1['antelen']>1)&(rule1.confidence>0.7)]
```

```python
rule2=association_rules(df_fre,metric='confidence',min_threshold=0.7)
rule2['antelen']=rule2.antecedents.apply(lambda x:len(x))
rule2=rule2[(rule2.antelen==1)&(rule2.lift>1)]
rule2
```

```python
rule2.plot(kind='scatter',x='support',c='r',
          y='confidence',s=30,figsize=(8,5))
plt.grid('on')
plt.xlabel('支持度',fontproperties=fonts,size=12)
plt.ylabel('置信度',fontproperties=fonts,size=12)
plt.title('38个规则散点图',fontproperties=fonts)
plt.show()
```

```python
import networkx as nx
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #显示中文标签
plt.rcParams['axes.unicode_minus']=False
plt.figure(figsize=(12,12))
G=nx.DiGraph()
for ii in rule2.index:
    G.add_edge(rule2.antecedents[ii],rule2.consequents[ii],weight=rule2.support[ii])

elarge=[(u,v) for(u,v,d)in G.edges(data=True)if d['weight']>0.6]
emiddle=[(u,v) for(u,v,d)in G.edges(data=True)if (d['weight']<=0.6)&(d['weight']>=0.45)]
esmall=[(u,v) for(u,v,d)in G.edges(data=True)if d['weight']<0.45]
pos=nx.circular_layout(G)
nx.draw_networkx_nodes(G,pos,alpha=0.4,node_size=rule2.confidence*500)
nx.draw_networkx_edges(G,pos,edgelist=elarge,
                        width=2,alpha=0.6,edge_color='r')
nx.draw_networkx_edges(G,pos,edgelist=emiddle,
                       width=2,alpha=0.6,edge_color='g')
nx.draw_networkx_edges(G,pos,edgelist=esmall,
                        width=2,alpha=0.6,edge_color='b')
nx.draw_networkx_labels(G,pos,font_size=10,fontproperties=fonts)
plt.axis('off')
plt.title('38个规则使用支持度连接',fontproperties=fonts)
plt.show()
```

```python
wine = datasets.load_wine(return_X_y=True)
wineX=wine[0]
wineX.shape
```

```python
pca=PCA(n_components=5)
wineNew=pca.fit_transform(wineX)
print('累计方差贡献率',np.sum(pca.explained_variance_ratio_))
```

```python
K=np.arange(1,20)
ser=[]
for ii in K:
    kmean =cluster.KMeans(n_clusters=ii,random_state=1)
    kmean.fit(wineNew)
    ser.append(kmean.inertia_)
plt.figure(figsize=(8,5))
plt.plot(K,ser,'r-o')
plt.xlabel('聚类数目',fontproperties=fonts,size=12)
plt.ylabel('类内误差平方和',fontproperties=fonts,size=12)
plt.title('K-means聚类',fontproperties=fonts)
plt.xticks(np.arange(1,20,2))
plt.grid('on')
plt.show()
```

```python
kmean=cluster.KMeans(n_clusters=3,random_state=1)
k_pre=kmean.fit_predict(wineNew)
fig=plt.figure(figsize=(6,5))
ax1=fig.add_subplot(111)
ax1.scatter(wineNew[k_pre==0,0],wineNew[k_pre==0,1],c='r',alpha=1,marker='*')
ax1.scatter(wineNew[k_pre==1,0],wineNew[k_pre==1,1],c='g',alpha=1,marker='s')
ax1.scatter(wineNew[k_pre==2,0],wineNew[k_pre==2,1],c='b',alpha=1,marker='d')
ax1.scatter(kmean.cluster_centers_[:,0],kmean.cluster_centers_[:,1],c='white',marker='o',s=80,edgecolor='k')
ax1.set_xlabel('Principal component 1')
ax1.set_ylabel('Principal component 2')

ax1.set_title('K means 聚类为3类',fontproperties=fonts)
plt.show()
```

```python
from sklearn.metrics import silhouette_score,silhouette_samples
sil_score=silhouette_score(wineNew,k_pre)
sil_samp_val=silhouette_samples(wineNew,k_pre)
plt.figure(figsize=(8,5))
y_lower=10
n_clu=len(np.unique(k_pre))
for ii in np.arange(n_clu):
    iiclu_sil_samp_sort=np.sort(sil_samp_val[k_pre==ii])
    iisize=len(iiclu_sil_samp_sort)
    y_upper =y_lower +iisize
    color=cm.Spectral(ii/n_clu)
    plt.fill_betweenx(np.arange(y_lower,y_upper),0,iiclu_sil_samp_sort,
                      facecolor =color,alpha=0.7)
    plt.text(-0.05,y_lower+0.5*iisize,str(ii))
    y_lower=y_upper+5
plt.axvline(x=sil_score,color='red',label='mean:'+str(np.round(sil_score,3)))
plt.xlim([-0.1,1])
plt.yticks([])
plt.legend(loc=1)
plt.xlabel('Silhouette Coefficient value')
plt.ylabel('Cluster label')
plt.title('Silhouette value for Kmeans cluster')
plt.show()
```

```python
import sklearn.decomposition as skd
from sklearn.feature_extraction.image import extract_patches_2d
from sklearn.feature_extraction.image import reconstruct_from_patches_2d
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.util import random_noise
```

```python
im=imread('data\莱娜.tiff')
orimga=rgb2gray(im)
mode='gaussian'
sigma=30
sigma2=sigma**2/(255**2)
noimga=random_noise(orimga,mode=mode,var=sigma2,seed=1243,clip=True)
print('noimga',noimga.min(),'~',noimga.max())
```

```python
patch_size=(10,10)
noimgadata=extract_patches_2d(noimga,patch_size)
noimgadata=noimgadata.reshape(noimgadata.shape[0],-1)
noimgadata=noimgadata.T
interp=noimgadata.mean(axis=0)
print('interp.shape:',interp.shape)
noimgadata=noimgadata-interp
noimgadata=noimgadata.T
print('训练模型数据shape is:',noimgadata.shape)
print('带噪声图像取值为:',noimgadata.min(),'~',noimgadata.max())
print('noimgadata.shape:',noimgadata.shape)
```

```python
def DCTdict(m,n):
    import numpy as np
    dictionary=np.zeros((m,n))
    for k in range(n):
        V=np.cos(np.arange(0,m)*np.pi*k/n)
        if k >0:
            V = V - np.mean(V)
        dictionary[:,k]=V/np.linalg.norm(V)
    dictionary = np.kron(dictionary,dictionary)
    dictionary = dictionary.dot(np.diag(np.sum((dictionary*dictionary),axis=0)))
    return dictionary
```

```python
m=16
n=patch_size[0]
dctdictionary=DCTdict(m,n)
n_components=m**2
n_iter=100
code,dictionary=skd.dict_learning_online(noimgadata,n_components=n_components,
                                        alpha=0.2,n_iter=n_iter,return_code=True,
                                        method='lars',n_jobs=6,dict_init=dctdictionary,random_state=1234)
print('dictionary shape=',dictionary.shape)
print('sparse code shape=',code.shape)

```

```python
import numpy as np
denoimgadata=np.dot(code,dictionary)
denoimgadata=denoimgadata.T
denoimgadata=denoimgadata+interp
denoimgadata=denoimgadata.T
patch=denoimgadata.reshape(len(denoimgadata),*patch_size)
denoimga=reconstruct_from_patches_2d(patch,orimga.shape)
print('denoimga.shape',denoimga.shape)
```
