```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
%config InlineBeckend.figure_format='retina'
from sklearn import metrics
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

sns.set_style("whitegrid")
import warnings
warnings.filterwarnings('ignore')
```

```python
#Try to train directly with raw data
train_x,test_x,train_y,test_y=train_test_split(
    diabetes.loc[:, diabetes.columns != 'Outcome'],
    diabetes['Outcome'], stratify=diabetes['Outcome'],
    random_state=50)
```

```python
#KNN algorithm is used to find the relationship between the prediction accuracy (Y-axis) and the number of adjacent points (x-axis) between the training set and the test set
'''
According to the image analysis, the accuracy of KNN model is about 78%, 
and there is over fitting when k < 5,
good training accuracy and test accuracy when k = 6
'''
training_accuracy=[]
test_accuracy=[]
neighbors_settings =range(1,11)
for n_neighbors in neighbors_settings:
    knn=KNeighborsClassifier(n_neighbors=n_neighbors)
    knn.fit(train_x,train_y)
    training_accuracy.append(knn.score(train_x,train_y))
    test_accuracy.append(knn.score(test_x,test_y))
plt.figure()
plt.plot(neighbors_settings,training_accuracy,label='training accuracy')
plt.plot(neighbors_settings,test_accuracy,label='test accuracy')
plt.ylabel('Accuracy')
plt.xlabel('n_neighbors')
plt.legend()
```

```python
knn=KNeighborsClassifier(n_neighbors=6)
knn.fit(train_x,train_y)
print('Accuracy of K-NN Classifier on train set %.2f'%knn.score(train_x,train_y))
print('Accuracy of K-NN Classifier on test set %.2f'%knn.score(test_x,test_y))
```

```python
# logistic regression analysis
"""
The accuracy of the model with regularization parameter C=1(default value) was 77%
the training set and 74% on the test set
"""
logreg = LogisticRegression(solver='liblinear').fit(train_x,train_y)
print("Accuracy of LogisticRegression on train set:%.2f"%logreg.score(train_x, train_y))
print("Accuracy of LogisticRegression on test set:%.2f"%logreg.score(test_x,test_y))
```

```python
# Remove outliers from skinthickness
max_skinthickness = diabetes2.SkinThickness.max()
diabetes2 = diabetes2[diabetes2.SkinThickness!=max_skinthickness]
```

```python
# 
def fill_zero(df, field, target):
    mean_by_target = df.loc[df[field] != 0, [field, target]].groupby(target).mean()
    diabetes2.loc[(df[field] == 0)&(df[target] == 0), field] = mean_by_target.iloc[0][0]
    diabetes2.loc[(df[field] == 0)&(df[target] == 1), field] = mean_by_target.iloc[1][0]

    # run the function
for col in ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']:   
    replace_zero(diabetes2, col, 'Outcome')  
```
